```python
training = training[['cons_12m', 'cons_gas_12m','margin_gross_pow_ele','net_margin','num_years_antig','churn']]
training.info()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=100)
for train_index, test_index in split.split(training, training["churn"]):
    train_set = training.loc[train_index]
    test_set = training.loc[test_index]   
X_train = train_set[['cons_12m', 'cons_gas_12m']]
y_train = train_set['churn']
y_train.value_counts()
```

```python
X_test = test_set[['cons_12m', 'cons_gas_12m']]
y_test = test_set['churn']
y_test.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state = 0)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y_train,X_train)
result=logit_model.fit()
print(result.summary())
```
