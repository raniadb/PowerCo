```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=100)

def split_dfmodel(df):
    for train_index, test_index in split.split(df, df['churn']):
        train_set = df.iloc[train_index]
        test_set = df.iloc[test_index]
    dfX_train = train_set.loc[:, train_set.columns != 'churn']
    dfy_train = train_set.iloc[:,-1]
    dfX_test = test_set.loc[:, test_set.columns != 'churn']
    dfy_test = test_set.iloc[:,-1]
    return dfX_train,dfX_test,dfy_train,dfy_test
```

```python
def allrows(ytest,yproba,yscore):
    df = pd.DataFrame(ytest.values,columns=['test'],index=ytest.index)
    df['proba0'] = yproba[:,0]
    df['proba1'] = yproba[:,1]
    df['forecast_bill_12m'] = training_data['forecast_bill_12m']
    df = df.dropna(how='any',axis=0)
    df = df.sort_values('proba1',ascending = False)
    return df
```

```python
def firstnrows(dfall,n,redu):
    df = dfall.iloc[:n,:].copy()
    df['actual_bill_12m'] = 0
    df.loc[df['test'] == 0,'actual_bill_12m'] = df['forecast_bill_12m']*(-redu)
    df.loc[df['test'] == 1,'actual_bill_12m'] = df['forecast_bill_12m']*(1-redu)
    tot = df['actual_bill_12m'].sum()
    return df,tot
```

```python
def print_threshold_precision_total(dfall,n,redu):
    df = dfall.copy()
    df['pred'] = 0
    df.iloc[:n,-1] = 1
    th=df.iloc[n-1,2]
    preci = precision_score(df['test'],df['pred'])
    df,tot = firstnrows(df,n,redu)
    return df,th,preci,tot
```

```python
dfallrows = allrows(y_test,y_proba,y_score)
for i in range(1,dfallrows.shape[0]+1):
#for i in range(90,91):
    dfn,threshold,precis,total = print_threshold_precision_total(dfallrows,i,0.20)
    if total > 90000:
        print(i,'Threshold {:.5f}'.format(threshold),'Precision: {0:0.2f}'.format(precis),'Total: {0:0.5f}'.format(total))
```

```python
data_plus_histdata_nums_mean_agg = data_plus_histdata_nums_mean.copy()
```

```python
pd.options.mode.chained_assignment = None
def calc_agg(df):
    for i in (1,2,3):
        sem1_var = [c for c in X_train.columns if ((c[:18] == 'price_p' + str(i) + '_var_2015-') & (c[-2:] in ['01','02','03','04','05','06']))]
        sem1_fix = [c for c in X_train.columns if ((c[:18] == 'price_p' + str(i) + '_fix_2015-') & (c[-2:] in ['01','02','03','04','05','06']))]
        sem2_var = [c for c in X_train.columns if ((c[:18] == 'price_p' + str(i) + '_var_2015-') & (c[-2:] in ['07','08','09','10','11','12']))]
        sem2_fix = [c for c in X_train.columns if ((c[:18] == 'price_p' + str(i) + '_fix_2015-') & (c[-2:] in ['07','08','09','10','11','12']))]
        df['price_p' + str(i) + '_var_sem1'] = df[sem1_var].mean(axis=1)
        df['price_p' + str(i) + '_var_sem2'] = df[sem2_var].mean(axis=1)
        df['price_p' + str(i) + '_fix_sem1'] = df[sem1_fix].mean(axis=1)
        df['price_p' + str(i) + '_fix_sem2'] = df[sem2_fix].mean(axis=1)
        df = df[[c for c in df.columns if c not in (sem1_var + sem2_var + sem1_fix + sem2_fix)]]
    return df

data_plus_histdata_nums_mean_agg = calc_agg(data_plus_histdata_nums_mean_agg)
```

```python
dates_cols = [c for c in X_train_z.columns if c.lower()[:6] == 'z_days']
cat_cols = [c for c in X_train_z.columns if c.lower()[:2] != 'z_']
ts_cols = [c for c in X_train_z.columns if c.lower()[:7] == 'z_price']
num_cols = list(set(X_train_z.columns)-set(dates_cols)-set(cat_cols)-set(ts_cols))
```

```python
bcg_input_cat = [Input(shape=(1,)) for c in cat_cols]
bcg_input_num = Input(shape=(len(num_cols+dates_cols),))
bcg_input_ts = Input(shape=(12,6))

bcg_output_cat = [Flatten()(Dropout(0.1)(Embedding(X_train_z[c].max()+2, 2)(x))) for c,x in zip(cat_cols, bcg_input_cat)]
bcg_output_num = Dropout(0.1)(Dense(100, activation='selu')(bcg_input_num))
bcg_output_ts = Bidirectional(LSTM(25, return_sequences=True, dropout= 0.1, recurrent_dropout= 0.1),
                              merge_mode='ave')(bcg_input_ts)
bcg_output_ts = LSTM(26, return_sequences=False, dropout= 0.1, recurrent_dropout= 0.1)(bcg_output_ts)

bcg_output = concatenate(bcg_output_cat + [bcg_output_num, bcg_output_ts])
bcg_output = Dense(100, activation='selu')(bcg_output)
bcg_output = Dropout(0.5)(bcg_output)
bcg_output = Dense(100, activation='selu')(bcg_output)
bcg_output = Dropout(0.1)(bcg_output)
bcg_output = Dense(2, activation='softmax')(bcg_output)
```

```python
bcg_model = Model(bcg_input_cat+[bcg_input_num, bcg_input_ts], bcg_output)
```

```python
bcg_model.summary()
```

```python
bcg_model.compile(optimizer=RMSprop(lr=0.001, rho=0.9), loss='binary_crossentropy', metrics=['accuracy'])
```
